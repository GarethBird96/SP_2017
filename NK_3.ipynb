```python
import gps_particle_data
import wget
import os
import numpy as np 
import time
import json
import sys
import math
from datetime import datetime, timedelta, date
from itertools import compress
from random import randint
import urllib2
import aacgmv2

import matplotlib
from matplotlib import gridspec
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Script simple plots for NK nuclear tests on the following dates:
#9 October 2006, 25 May 2009, 12 February 2013, 6 January 2016, 9 September 2016

start_date = datetime(2006,9,28,0,0,0);
end_date = datetime(2006,10,30,0,0,0);
ms = gps_particle_data.meta_search();
ms.load_data(start_date,end_date);
```

```python
output_data = ms.get_all_data_by_satellite();
for ns in output_data.keys():
    print 'SATELLITE', ns, '\n'
    x =  output_data[ns]['L_shell']
    el = output_data[ns]['rate_electron_measured']
    pr = output_data[ns]['rate_proton_measured']
    day = output_data[ns]['decimal_day']
    fig = plt.figure(figsize=(10, 8), dpi=120)
    gs = gridspec.GridSpec(3,1)
    ax0 = plt.subplot(gs[0])
    #plt.xlim(278, 286)
    #plt.ylim(0,3000)
    plt.ylabel('Electron rate / Hz')

    labels = ['0.14-0.23', '0.23-0.41', '0.41-0.77', '0.77-1.25', '1.26-68', '1.3-1.7', '1.7-2.2', '2.2-3.0', '3.0-4.1', '4.1-5.8', '>5.8']
    colour = ['rosybrown', 'blue', 'red', 'green', 'magenta', 'navy', 'aqua', 'orange', 'yellow' , 'lime', 'peru', 'slategrey', 'teal', 'maroon', 'mediumspringgreen']
    y = np.split(el, range(len(el[0])), axis = 1)
    for i in range(len(el[0])):    
        ax0.scatter(day, y[i+1], color = colour[i], label = labels[i])
    ax0.axvline(x=282, color = 'k')
    plt.legend(fontsize = 'xx-small')

    ax1 = plt.subplot(gs[1], sharex = ax0)
    labels = ['6.0-10', '10-50', '16-128', '57-75', '>75']
    colour = ['rosybrown', 'blue', 'red', 'green', 'magenta', 'navy', 'aqua', 'orange', 'yellow' , 'lime', 'peru', 'slategrey', 'teal', 'maroon', 'mediumspringgreen']
    p = np.split(pr, range(len(pr[0])), axis = 1)
    for i in range(len(pr[0])):    
        ax1.scatter(day, p[i+1], color = colour[i], label = labels[i])
    ax1.axvline(x=282, color = 'k')
    plt.ylabel('Proton rate / Hz')
    #plt.xlim(278, 286)
    plt.legend(fontsize = 'xx-small')

    ax2 = plt.subplot(gs[2], sharex = ax0)
    ax2.scatter(day, x)
    ax2.axvline(x=282, color = 'k')
    plt.ylim(4,80)
    #plt.xlim(278, 286)
    plt.ylabel('L-shell')
    plt.xlabel('decimal day')
    plt.subplots_adjust(hspace = .0)
    plt.show()
```

```python
#Normalised cross correlation values as compared to satellite 53.

for ns in output_data.keys():
    if len(output_data[ns]) !=0 and ns != 53:
        print 'SATELLITE', ns, '\n', '===============\n'
        for i in range(11):
            a0 = np.asarray(output_data[53][signal])[:,i]
            ar2 = np.asarray(output_data[ns][signal])[:,i]
            cor = np.correlate(a0, ar2)/(math.sqrt(np.sum(a0**2)*np.sum(ar2**2)))
            print 'channel:', i, 'X-cor  --->   ', cor[0]
        print'\n'
```
